```
import cmath  # libreria para números complejos
import math
import matplotlib.pyplot as plt
import numpy as np
```

```
def classical_method(a,b,c): #esta función calculará las raices para polinómios de orden 2 mediante el metodo clásico
  disc=b**2-(4*a*c) #discriminante
  
  if disc>=0: 
    root1=((-b)+math.sqrt(disc))/(2*a)
    root2=((-b)-math.sqrt(disc))/(2*a)
  else:
    root1=(-b+cmath.sqrt(disc))/(2*a)
    root2=(-b-cmath.sqrt(disc))/(2*a)
  print("Método clásico")
  print("raiz 1: {0} , raíz 2: {1}".format(root1,root2))
```

```
def Po_Shen_Loh_method(a,b,c):#función que calcula las raices de un polinomió de orden 2 , mediante el nuevo método Po Shen Loh

  if a!=1: #si a!=1 normalizamos la ecuación para usar el método
    inv=1/a  
    b*=inv
    c*=inv

  sum_flag=-b/2           
  sum_flag_2=(b**2)/4
  u_2=sum_flag_2-c #viene a ser una especie de discriminante para determinar si es una raiz compleja o no

  if u_2<0:
    root1=sum_flag+cmath.sqrt(u_2)
    root2=sum_flag-cmath.sqrt(u_2)
  else:
    root1=sum_flag+math.sqrt(u_2)
    root2=sum_flag-math.sqrt(u_2)
  print("Método Po Shen Loh ")
  print("raiz 1: {0} , raíz 2: {1}".format(root1,root2))
  return root1,root2

```

###3. to plot the newton fractals for the corresponding complex quadratic polynomials

```
def newton_set1(f,df,roots,z_c_r=0, z_c_i=0, zh = 5, zw=5, width=500, height=500, zoom=1,  niter=256,m=1):
    """ Fractals using newton-raphson """

    # Pixels array
    pixels = np.arange(width*height*3, dtype=np.uint32).reshape(height, width, 3)

    h = 1e-7 # step size for numerical derivative
    eps = 1e-3 # max error allowed

    # Bounding roots
   

    # Color multiplication factor
    # NOTE: Increasing this darkens the image and lightens it otherwise
    multcol = [[5,0,0],[0,5,0],[0,0,5]]
    r=[255,0,0]
    g=[0,255,0]
    b=[0,0,255]
    for y in range(height):
            zy =   z_c_i+(((y-(height/2))/height)*zh*zoom)

            for x in range(width):
                zx = z_c_r+ (((x-(width/2))/width)*zw*zoom)

                # Mapping real -> complex plane
                z = complex(zx, zy)
                
                count = 0
                
                for i in range(niter):
                    # complex numerical derivative
                    dz = df(z)
                    if dz == 0:
                        
                        break

                    count += 1
                    if count > 255:
                      
                      break
                   
                    znext = z - m*f(z) / dz # Newton iteration   m es la multiplicidad de la raiz, de esta manera converge más rápido

                    if abs(znext - z) < eps: # stop when close enough to any root
    
                        break
                
                    z = znext

                # Pixels colored using the roots
                flag=True
                
                for i,val in enumerate(roots):
                  
                  if abs(z-val)<=eps:
                      # color red
                      
                      pixels[height-y-1,x] = (r[i] - count*multcol[i][0], g[i]- count*multcol[i][1], b[i]- count*multcol[i][2])
                      flag=False
                      break
                  
                  
                if flag:   
                    pixels[height-y-1,x] = (0, 0, 0)
 
    return pixels

def display1(name,f,df,roots,z_c_r=0, z_c_i=0, zh = 1, zw=1, width=500, height=500, zoom=1,  niter=256,m=1):
    """ Display a newton-raphson fractal """
    print("Z region from (",z_c_r+ (((0-(width/2))/width)*zw*zoom),",",z_c_i+(((0-(height/2))/height)*zh*zoom) ,") to (",z_c_r+ (((width-(width/2))/width)*zw*zoom),",",z_c_i+(((height-(height/2))/height)*zh*zoom),")")
    pimg = newton_set1(f,df,roots,z_c_r=z_c_r, z_c_i=z_c_i, zh = zh, zw=zw, width=width, height=width, zoom=zoom,  niter=niter,m=m)
    plt.title('Newton Fractal f(z)='+name)
    plt.xlabel("real"+"("+ str(z_c_r+ (((0-(width/2))/width)*zw*zoom))+","+str(z_c_r+ (((width-(width/2))/width)*zw*zoom))+")")
    plt.ylabel("imaginary"+"("+str(z_c_i+(((0-(height/2))/height)*zh*zoom))+","+str(z_c_i+(((height-(height/2))/height)*zh*zoom)) +")")
    plt.imshow(pimg)
    plt.show()
```

### 4. Plot the Newton iterates for points on the positive and negative axes for the complex quadratic polynomials in 3. Give an interpretation of the results.

- $x^2$

```
def eq1(x):
  return x**2

def deq1(x):
  return 2*x
```

```
def plot_iteration(z,eq1,deq1,title):
  r=[]
  i=[]
  for k in range(100):
    dz = deq1(z)
    z = z - eq1(z) / dz # Newton iteration
    r.append(np.real(z))
    i.append(np.imag(z))
  plt.title('Newton iterates form '+title)
  plt.plot(r)
  plt.ylabel('Real part')
  plt.show()
  plt.plot(i)
  plt.ylabel('Imagnary part')
  plt.show()
```

```
plot_iteration(-10+0.0j,eq1,deq1,"-10+0.0j")
plot_iteration(10+0.0j,eq1,deq1,"10+0.0j")
plot_iteration(0+10.0j,eq1,deq1,"0+10.0j")
plot_iteration(0-10.0j,eq1,deq1,"0-10.0j")
plot_iteration(1+0j,eq1,deq1,"0+1j")
```

####Análisis
Cuando nos encontramos en el eje positivo o negativo de la parte real, lo que ocurre es que la parte imaginaria se mantiene en 0, y la parte real inicia a converger hacia su raíz en este caso es cero, aproximadamente converge en 5 pasos. por otro lado cuando nos encontramos en el eje positivo o negativo de la parte imaginaria, lo que ocurre es que la parte real se queda en cero, mientras la parte imaginaria inicia a converger hacia cero, la cual es la raíz. En este caso si tomamos puntos en los ejes la convergencia ocurre relativamente rapido aproximadamente en 5 o 6 pasos
